```python
# Dependencies

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from config import spotify_id
from config import spotify_secret
import json
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd
import sys
import pprint



#Store spotify authorization as variable
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=spotify_id,
                                                           client_secret=spotify_secret))


```

```python
# Create list of favorite artists
artists = ['j balvin', 'ozuna', "chris lake", "carlos vives"]

# Create empty lists 
idx_list = []
tracks = []
artists_list = []
albums = []
release_dates = []
preview_url = []
popularity = []
duration = []



# Use Spotify API through Spotipy to pull desired info
for artist in artists:
    results = sp.search(q=artist, limit=10)
    for idx, track in enumerate(results['tracks']['items']):
    
        idx_list.append(idx)
        artists_list.append(artist)
        tracks.append(track['name'])
        preview_url.append(track['preview_url'])
        popularity.append(track['popularity'])
        duration.append(track['duration_ms'])
        release_dates.append(track['album']['release_date'])
 



# Create 

dict = {'Artist': artists_list, 'Track': tracks, "Duration":duration,"Release Date":release_dates ,"Popularity" : popularity,'Preview URL': preview_url }  


spotify_df = pd.DataFrame(dict) 


spotify_clean = spotify_df.drop_duplicates(subset=['Track'])
spotify_clean.head(50)
```
