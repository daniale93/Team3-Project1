```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file = "../Resources/data.csv"

spotify_df = pd.read_csv(file)
spotify_df.head()
```

```python
#new data frame of 200 samples
song_samples_df= spotify_df.sample(n=200)
```

```python
#Danceability to Popularity

song_samples_df.plot(kind= "scatter", x= "danceability", y= "popularity")
plt.title("Sample of Danceability to Popularity")
plt.xlabel("Danceability")
plt.ylabel("Popularity")
plt.savefig("../PopFigures/Danceability.png")
plt.show()
```

```python
#Energy to Popularity

song_samples_df.plot(kind= "scatter", x= "energy", y= "popularity")
plt.title("Sample of Energy to Popularity")
plt.xlabel("Energy")
plt.ylabel("Popularity")
plt.savefig("../PopFigures/Energy.png")
```

```python
#duration to Popularity

song_samples_df.plot(kind= "scatter", x= "duration_ms", y= "popularity")
plt.title("Sample Duration to Popularity")
plt.xlabel("Duration in ms")
plt.ylabel("Popularity")
plt.savefig("../PopFigures/SampleDuration.png")
```

```python
#FULL DATA: duration to Popularity

spotify_df.plot(kind= "scatter", x= "duration_ms", y= "popularity")

plt.title("Full Data of Duration to Popularity")
plt.xlabel("Duration in ms")
plt.ylabel("Popularity")
plt.savefig("../PopFigures/FullDataDuration.png")
```

```python
#FULL DATA- half view: duration to Popularity

spotify_df.plot(kind= "scatter", x= "duration_ms", y= "popularity")
# The x limits of our scatter plot is 0 to 500,000
plt.xlim(0, 500000)
plt.ylim(50,110)
plt.title("Looking at the Peak of Duration to Popularity")
plt.xlabel("Duration in ms")
plt.ylabel("Popularity")
plt.savefig("../PopFigures/PeakDuration.png")
```
