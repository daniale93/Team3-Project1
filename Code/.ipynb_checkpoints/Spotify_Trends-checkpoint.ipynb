```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn import datasets

file = "../Resources/data.csv"

spotify_df = pd.read_csv(file)
spotify_df.head()
```

```python
# Create dataframes with year averages
spotify_yr_inst = spotify_df.groupby("year")["instrumentalness"].mean().reset_index()
spotify_yr_live = spotify_df.groupby("year")["liveness"].mean().reset_index()
spotify_yr_aco = spotify_df.groupby("year")["acousticness"].mean().reset_index()
spotify_yr_dance = spotify_df.groupby("year")["danceability"].mean().reset_index()
spotify_yr_energy = spotify_df.groupby("year")["energy"].mean().reset_index()
spotify_yr_spe = spotify_df.groupby("year")["speechiness"].mean().reset_index()
spotify_yr_bpm = spotify_df.groupby("year")["tempo"].mean().reset_index()
spotify_yr_val = spotify_df.groupby("year")["valence"].mean().reset_index()
spotify_yr_dur = spotify_df.groupby("year")["duration_ms"].mean().reset_index()

spotify_yr_inst
```

```python
x_values = spotify_yr_inst['year']
y_values = spotify_yr_inst['instrumentalness']
plt.plot(x_values,y_values )
plt.title("Instrumentalness over time")
plt.xlabel("Year")
plt.ylabel("Instrumentalness")


(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_values,regress_values,"r-")
plt.savefig("../TrendFigures/instrumentalness.png")
plt.show()
```
